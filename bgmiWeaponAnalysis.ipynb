### Clearing null values 

```python
weapons.dropna(axis=0,inplace=True)
```

```python
weapons
```

### Classfication on basis of weapon types

```python
weapons.groupby('Weapon Type').mean()
```

```python
weapons.groupby('Weapon Type').count()
```

### Frequency analysis of Weapon Types

```python
plt.figure(figsize=(20,6))
plt.hist(weapons['Weapon Type'],bins=20,color='Orange')
plt.xlabel('Weapon Type')
plt.ylabel('Frequency')
plt.title('Most Available Guns')
```

### Weapon Types vs Firing Range Analysis

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Weapon Type', y='Range',data=pd.read_csv('weapons.csv'))
plt.title('Weapon Type vs Range')
plt.grid('true')
```

### Headshot Damage vs Range Analysis

```python
plt.figure(figsize=(15,5))
plt.bar(weapons['Weapon Type'], weapons['Range'], label='Range', color='yellow')
plt.bar(weapons['Weapon Type'], weapons['HDMG_0'], label='Headshot Damage', color='blue')
plt.xlabel=("Weapon Type")
plt.ylabel=("Range & Headshot Damage")
plt.title=("Headshot Damage vs Range Analysis")
plt.grid("true")
plt.legend()
```

### Damage Analysis wrt Types of Weapons

```python
weapondamage=weapons.groupby('Weapon Type')['Damage'].mean().reset_index()

plt.figure(figsize=(15,5))
sns.barplot(y='Weapon Type',x='Damage',data=weapondamage)
plt.grid('true')
```

### Range Analysis wrt Types of Weapons

```python
weaponrange=weapons.groupby('Weapon Type')['Range'].mean().reset_index()

plt.figure(figsize=(15,5))
sns.barplot(y='Weapon Type',x='Range',data=weaponrange)
plt.grid('true')
```

### Bullet Speed Analysis wrt Types of Weapons

```python
bulletspeed=weapons.groupby('Weapon Type')['Bullet Speed'].mean().reset_index()

plt.figure(figsize=(15,5))
sns.barplot(y='Weapon Type',x='Bullet Speed',data=bulletspeed)
plt.grid('true')
```

### Firing Rate Analysis wrt Types of Weapons

```python
Firerate=weapons.groupby('Weapon Type')['Rate of Fire'].mean().reset_index()

plt.figure(figsize=(15,5))
sns.barplot(y='Weapon Type',x='Rate of Fire',data=Firerate)
plt.grid('true')
```
